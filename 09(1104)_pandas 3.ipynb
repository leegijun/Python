# 함수 적용하기

```python
air.groupby(['PRODUCT_NBR', 'HOST1'])['ONAIR_MINS'].sum()
```

```python
host = air.groupby(['PRODUCT_NBR', 'HOST1'])['ONAIR_MINS'].sum()
```

```python
host.groupby('PRODUCT_NBR').apply(lambda x: x / x.sum())
```

```python
min_sums.groupby(level=0).apply(lambda x: x / x.sum())
```

# 피벗테이블
