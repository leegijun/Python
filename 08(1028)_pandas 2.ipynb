```python
def over_1_hour(group):
    return group['ONAIR_MINS'].sum() > 60
```

```python
grouped.filter(over_1_hour)
```

#### 방송횟수가 5회 미만인 제품 그룹

```python
grouped.filter(lambda group : group['ONAIR_MINS'].count() < 5)
```

#### 방송시간 합계가 60분 이상인 제품 그룹

```python
grouped.filter(lambda group : group['ONAIR_MINS'].sum() > 60)
```

##### 2015년도 3월 10일 이후에 첫방송한 제품

```python
grouped.filter(lambda group : group['ONAIR_DATE'].min() > '2015-03-10')
```

#### 제품 중에 첫방송 날짜가 3월 10일 이후

1. groupby 제품번호별 / 2. agg 첫방송날짜 / 3. >2015-03-10

```python
air_date = air.groupby('PRODUCT_NBR').agg({'ONAIR_DATE': min})
```

```python
air_date[air_date['ONAIR_DATE'] > '2015-03-15']
```

# Summarizing & Grouping 

```python
import pandas as pd
```

```python
air = pd.read_csv('data/product_airtime.csv')
```

```python
air.head()
```

```python
air.groupby('PRODUCT_NBR')['ONAIR_MINS'].sum()
```

```python
air.groupby('PRODUCT_NBR')['ONAIR_MINS'].sum().sort_values()
```

```python
min_sums = air.groupby(['PRODUCT_NBR', 'ONAIR_DATE'])['ONAIR_MINS'].sum()
```

```python
min_sums
```

```python
grouped = air.groupby('PRODUCT_NBR')
```

```python
grouped.get_group('P150000002')
```

```python
import numpy
```

```python
grouped['ONAIR_MINS'].agg([numpy.sum, numpy.mean, numpy.std])
```

```python
grouped.agg({'ONAIR_DATE': min,
             'ONAIR_MINS': [numpy.sum, numpy.mean, numpy.std]})
```

```python
def over_1_hour(group):
    return group['ONAIR_MINS'].sum() > 60
```

```python
grouped.filter(over_1_hour)
```

# lambda 함수 

```python
grouped.filter(lambda group: group['ONAIR_MINS'].sum() > 60)
```

## groupby level 

```python
min_sums
```

```python
min_sums.groupby(by='PRODUCT_NBR')
```

```python
min_sums.groupby(level=0)
```

```python
min_sums.groupby(level=0).agg(numpy.sum)
```

```python
min_sums.groupby(level=0).sum()
```

## apply

```python
min_sums.groupby(level=0).apply(lambda x: x / x.sum())
```

##  reset_index

```python
min_sums.reset_index()
```

## Pivot table

```python
pvt = pd.pivot_table(
    air,
    columns='ONAIR_DATE',
    index='PRODUCT_NBR',
    values='ONAIR_MINS',
    aggfunc=numpy.sum)
```

```python
pvt.head(10)
```

```python
pvtz = pvt.fillna(0)
```

```python
pvtz.head(10)
```

```python
pd.pivot_table(
    air,
    columns='ONAIR_DATE',
    index='PRODUCT_NBR',
    values='ONAIR_MINS',
    aggfunc=numpy.sum,
    fill_value=0)  # 결측값은 fill_value로 채운다
```
